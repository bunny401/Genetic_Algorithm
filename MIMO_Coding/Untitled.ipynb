```python
import matplotlib.pyplot as plt
import numpy as np

# Function to evaluate the formula: M * log2(M)
def evaluate_formula1(M):
    return M * np.log2(M)

# Function to evaluate the formula: (M/2) * log2(M)
def evaluate_formula2(M):
    return (M/2) * (np.log(M) / np.log(2))

# Set the value of M
M = 1000000

# Generate M values from 1 to M with fewer data points
num_points = 1000
M_values = np.linspace(1, M, num_points)

# Evaluate the formulas for each M value
y1 = evaluate_formula1(M_values)
y2 = evaluate_formula2(M_values)

# Filter out non-positive y values
mask = y1 > 0
M_values = M_values[mask]
y1 = y1[mask]
y2 = y2[mask]

# Calculate the number of functions to plot
num_functions = 2

# Calculate the figure size dynamically based on the number of functions
fig_width = 8 + num_functions * 2
fig_height = 6 + num_functions * 2

# Increase figure size
plt.figure(figsize=(fig_width, fig_height))

# Plot the graph for the first function
plt.plot(M_values, y1, label='M * log2(M)')

# Plot the graph for the second function
plt.plot(M_values, y2, label='(M/2) * log2(T) * (log2(M) / log2(3))')

plt.xlabel('M')  # X-axis label
plt.ylabel('Output')
plt.title('Graphs of Functions')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.xticks(np.arange(0, M+1, 80000))  # Adjust x-axis tick values
plt.ylim(bottom=1)  # Set the minimum y-value to 1
plt.grid(True)
plt.legend()
plt.show()
```
